**In the data set we two types of data**

- Numerical Data

**Categorical data**

```python
sales_df.select_dtypes(include='object').head(2)
```

```python
cate_data=sales_df.select_dtypes(include='object')
cate_data.head(2)
```

```python
cate_data.drop(['Cust_name','Product_ID'],axis=1,inplace=True)
```

```python
col=cate_data.select_dtypes(include='object').columns.unique()
col
```

**Categorical data analysis and visualization**

```python
plt.figure(figsize=(38,15))
for idx, col in enumerate(col, start=1):
    
    
    plt.subplot(2, 3, idx)
    category_counts = cate_data[col].value_counts()
    sns.countplot(cate_data,x=col,order=category_counts.keys(),color='orange')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=27)
    plt.savefig("Sales_CategoricalPlot.png") 
plt.tight_layout
plt.show()
```

<h1 style="font-size:40px;,color='Blue'">EXPLORATORY DATA ANALYSIS</h1>

**Relation between Gender and Amount**

```python
sales_df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

**we use groupby function to determine total Amount Spending particular gender**

```python
sales_gen=sales_df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,legend='auto')
plt.savefig("Gender_vs_Amount_Plot.png")
plt.show()
```

```python
ax = sns.barplot(data = sales_gen, x = 'Gender' , hue = 'Amount' , legend='auto')
for bars in ax.containers:
    ax.bar_label(bars)
```

**CONCLUSION**

- From Above graph we can see that most of the buyers are females and even the purchasing power of female are greater than men

**Relation Between Age Group and gender**

```python
ax = sns.countplot(data = sales_df, x = 'Age Group' , hue = 'Gender' , legend='auto')
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Age_vs_Genderplot.png")
```

**Relation between Total Amount VS Age Group**

```python
sales_age = sales_df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax = sns.barplot(x = 'Age Group', y = 'Amount' , data = sales_age,legend='auto')
#for bars in ax.containers:
    #ax.bar_label(bars)
```

**Conclusion**

- From Above graph we can see that most of the buyers are of age group between 26-35 yrs 

**Relation between State and Order**

```python
#total numbers of orders placed from  state
```

```python
sales_state = sales_df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state, x='State',y='Orders')
plt.savefig("ORDER_vs_STATE.png")
plt.show()
```

**Top 10 State where high order placed**

```python
sales_state = sales_df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state, x='State',y='Orders')
plt.savefig("Top_10.png")
plt.show()
```

**Conclusion**

- From above graph we can see that the uttar prades,maharashtra,karnataks,delhi etc are top state state where higher order paced

```python
#total amount/sales from top 10 state
```

```python
sales_state = sales_df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = sales_state, x='State',y='Amount')
plt.savefig("Amount_vs_STATE.png")
plt.show()
```

**Martial Status**

**Relation between Marital status,Gender and amount**

**Conclusion**

**Occupation**

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = sales_df, x = 'Occupation')
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("Occupation.png")
```

**Product Category**

```python
sns.set(rc={'figure.figsize':(25,10)})
ax = sns.countplot(data = sales_df, x = 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
plt.savefig("product_category.png")
```

```python
sales_state = sales_df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(38,10)})
sns.barplot(data = sales_state, x = 'Product_Category', y='Amount')
plt.savefig("ProductCtegory_vs_Amount.png")
plt.show()
```
